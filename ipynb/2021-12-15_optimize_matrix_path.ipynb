```python
### Referenced hyper-neutrino code on GitHub

# Loaded required packages
from collections import Counter
import numpy as np
from collections import defaultdict, deque

# Load input data
file_path = "C:\\Users\\JoeRatterman\\Documents\\GitHub\\2021-advent-of-code\\input-files\\2021-12-15_input.txt"
data_list = open(file_path, 'r').read().split('\n')

data_list
```

```python
# Function to split string
def str_split(word):
    return list(word)
```

```python
# Define function to check if matrix index is valid
def is_valid(array, index):    
    row, column = index
    return 0 <= row < len(array) and 0 <= column < len(array[row])

def get_next(node, offset):
    """Find the next location based on an offset from the current location."""
    row, column = node
    row_offset, column_offset = offset
    return row + row_offset, column + column_offset
```

```python
# Function to build matrix graph
def generate_graph(array, list_of_points, neighbors):

    # Create empty dictionary 
    graph_dict = defaultdict(list)

    # Loop through each element in graph
    for q in range(len(list_of_points)):

        # Loop through all neighboring points
        for i in range(len(neighbors)):
            next_spot = get_next(list_of_points[q], neighbors[i])

            # If point is valid add to dictionary
            if is_valid(array, next_spot):
                graph_dict[list_of_points[q]].append(next_spot)

    return(graph_dict)

    
```

```python
# Convert data input to matrix
for i in range(len(data_list)):

    if i == 0:
        input_matrix = np.array(str_split(data_list[i]))
    else:
        input_matrix = np.vstack([str_split(data_list[i]), input_matrix])

# Convert values to integers
#input_matrix = [list(map(int, i)) for i in input_matrix]
input_matrix = np.flipud(input_matrix).astype(int)
#input_matrix = np.asarray(input_matrix).astype(int)
input_matrix
```

```python
# Create list of tuples
tuple_list = []
for i in range(len(data_list)):
    for j in range(len(data_list[0])):
        tuple_list.append((i,j))

tuple_list[0:5]
```

```python
 # Locations to check for each spot
pair_check = ((0,1), (0,-1), (1,0), (-1, 0))
pair_check
```

```python
# Create dictionary of all touching points
point_graph = generate_graph(array = input_matrix, list_of_points = tuple_list, neighbors = pair_check)
point_graph[(9,9)]
```

```python
# Create cost dictionary 
cost_dict = {}
for i in range(len(tuple_list)):
    cost_dict[tuple_list[i]] = input_matrix[tuple_list[i]]

cost_dict[(9,8)]
```

```python
# Initialize variables
visited_list = []
path_list = [(0, 0, 0)]

# Loop through until reaching end of matrix
while True:

    # Pull values for lowest risk factor in list
    rf, x, y = path_list.pop(path_list.index(min(path_list, key = lambda t: t[0])))

    # Loop stop criteria
    if (x, y) in visited_list: continue
    if (x, y) == (len(input_matrix) - 1, len(input_matrix[x]) - 1):
        print(rf)
        break

    # Append visited tuple to list
    visited_list.append((x, y))

    # For each location visited calculate risk factor for adjacent cells
    for new_x, new_y in point_graph[(x, y)]:

        path_list.append((rf + cost_dict[(new_x, new_y)], new_x, new_y))

    
```
