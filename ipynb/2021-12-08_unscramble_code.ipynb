```python
# Load required packages
import numpy as np

# Load input data
file_path = "C:\\Users\\JoeRatterman\\Documents\\GitHub\\2021-advent-of-code\\input-files\\2021-12-08_input.txt"
data_list = open(file_path, 'r').read().split('\n')


# Dictionary for number & count of characters involved
count_dict = {
    "0": 6,
    "1": 2,
    "2": 5,
    "3": 5,
    "4": 4,
    "5": 5,
    "6": 6,
    "7": 3,
    "8": 7,
    "9": 6
}


# Dictionary of what letter combinations should make up each number
letter_dict = {
    "0": "abcefg",
    "1": "cf",
    "2": "acdeg",
    "3": "acdfg",
    "4": "bcdf",
    "5": "abdfg",
    "6": "abdefg",
    "7": "acf",
    "8": "abcdefg",
    "9": "abcdfg"
}

```

```python
# Function to split string
def str_split(word):
    return list(word)
```

```python
# Lists to test bearing_calc function
test_list = ['be', 'abcdefg', 'bcdefg', 'acdefg', 'bceg', 'cdefg', 'abdefg', 'bcdef', 'abcdf', 'bde', 'abcdefg', 'bcdef', 'bcdefg', 'bceg']
test_list = [''.join(sorted(s)) for s in test_list]
```

```python
# Function to convert list of letter combinations to correct letters 
# Used in phase 2
# Bit of a brute force approach
def bearing_calc(test_list):

    # Dictionary to store results
    line_dict = {}

    for i in range(len(test_list)):

        # Assign # 1
        if len(test_list[i]) == 2:
            line_dict[test_list[i]] = 1

        # Assign # 4
        if len(test_list[i]) == 4:
            line_dict[test_list[i]] = 4

    # Assign # 7
        if len(test_list[i]) == 3:
            line_dict[test_list[i]] = 7

        # Assign # 8
        if len(test_list[i]) == 7:
            line_dict[test_list[i]] = 8


    ### Calculate len(6) strings

    # Key & value lists for indexing
    key_list = list(line_dict.keys())
    val_list = list(line_dict.values())

    char_6 = list(filter(lambda x: len(x) == 6, test_list))
    char_6 = list(np.unique(char_6))

    # Calculate 9
    substring = str_split(key_list[val_list.index(1)])
    val_9 = [ f for f in char_6 if all(c in f for c in substring)]

    substring = str_split(key_list[val_list.index(4)])
    val_9 = [ f for f in char_6 if all(c in f for c in substring)]

    if len(val_9) > 0:
        line_dict[val_9[0]] = 9

   # Calculate 0
    substring = str_split(key_list[val_list.index(1)])
    val_0 = [ f for f in char_6 if all(c in f for c in substring)]

    char_6 = val_0
    substring = str_split(key_list[val_list.index(4)])
    val_0 = [ f for f in char_6 if not all(c in f for c in substring)]


    if len(val_0) > 0:
        line_dict[val_0[0]] = 0

    char_6 = list(filter(lambda x: len(x) == 6, test_list))
    char_6 = list(np.unique(char_6))

    # Calculate 6
    val_6 = list(filter(lambda x: not x in line_dict.keys(), char_6))

    if len(val_6) > 0:
        line_dict[val_6[0]] = 6



    ### Calculate len(5) strings

    # Key & value lists for indexing
    key_list = list(line_dict.keys())
    val_list = list(line_dict.values())

    char_5 = list(filter(lambda x: len(x) == 5, test_list))
    char_5 = list(np.unique(char_5))

    # Calculate 3
    substring = str_split(key_list[val_list.index(1)])
    val_3 = [ f for f in char_5 if all(c in f for c in substring)]

    if len(val_3) > 0:
        line_dict[val_3[0]] = 3

    # Calculate 5
    char_5 = list(filter(lambda x: x != val_3[0], char_5))
    char_5 = list(np.unique(char_5))


    substring = str_split(char_5[0])
    val_5 = [ f for f in [key_list[val_list.index(9)]] if all(c in f for c in substring)]

    if len(val_5) > 0:
        line_dict[char_5[0]] = 5


    if len(char_5) > 1:
        substring = str_split(char_5[1])
        val_5 = [ f for f in [key_list[val_list.index(9)]] if all(c in f for c in substring)]

        if len(val_5) > 0:
            line_dict[char_5[1]] = 5

    if len(char_5) > 2:
        
        substring = str_split(char_5[2])
        val_5 = [f for f in [key_list[val_list.index(9)]] if all(c in f for c in substring)]

        
        if len(val_5) > 0:
            line_dict[char_5[2]] = 5


    # Calculate 2
    val_2 = list(filter(lambda x: not x in line_dict.keys(), char_5))

    if len(val_2) > 0:
        line_dict[val_2[0]] = 2

    filtered_dict = {k:v for (k,v) in line_dict.items() if k in test_list[-4:]}
    # Key & value lists for indexing
    key_list = list(filtered_dict.keys())
    val_list = list(filtered_dict.values())

    str_obj = ""
    for i in test_list[-4:]:
        str_obj += str([val_list[key_list.index(i)]][0])

    # Return output number
    return(str_obj)
    



```
