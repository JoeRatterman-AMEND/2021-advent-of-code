```python
# Load required packages
import numpy as np

# Load input data
file_path = "C:\\Users\\JoeRatterman\\Documents\\GitHub\\2021-advent-of-code\\input-files\\2021-12-10_input.txt"
data_list = open(file_path, 'r').read().split('\n')


data_list
```

```python
# Function to split string
def str_split(word):
    return list(word)

# Function to find middle index of list
def findMiddle(input_list):
    middle = float(len(input_list))/2
    if middle % 2 != 0:
        return input_list[int(middle - .5)]
    else:
        return (input_list[int(middle)], input_list[int(middle-1)])
```

```python
# Identify valid characters
valid_open = ["(", "[", "{", "<"]
valid_close = [")", "]", "}", ">"]

# Dictionary to match close & open characters
valid_dict = {
        ")":"(", 
        "]":"[",
        "}":"{", 
        ">":"<"
}

# Dictionary to match close & open characters
close_dict = {
        "(":1, 
        "[":2,
        "{":3, 
        "<":4
}

# Dictionary for penalty terms
penalty_dict = {
        ")":3,
        "]":57,
        "}":1197,
        ">":25137
}
```

```python
# Function to calculate penalties for corruped lines
def character_check(str_obj):

    # Loop through each character object
    check_list = []
    for i in range(len(str_obj)):
        
        # Track characters tested
        check_list.append(str_obj[i])

        # Check if item is open or close
        # If open go to next iteration
        if valid_open.count(str_obj[i]) > 0:
           continue

        else:
            
            # If a close character check if it closes a chunk
            if valid_dict[str_obj[i]] == check_list[-2]:

                # If yes, remove the last two characters
                del check_list[-1]
                del check_list[-1]
        
            else:
                # If it doesn't close a chunk, line is corrupted. 
                # Return penalty value
                return(penalty_dict[str_obj[i]])
```

```python
# Loop through each line & calculate penalty values from each line
result_list = []
incomplete_index = []
for m in range(len(data_list)):

    val = character_check(str_split(data_list[m]))

    if val != None:
        result_list.append(val)
    else:
        incomplete_index.append(m)



print(incomplete_index)
sum(result_list)

```

```python
# Filter data input for only the incomplete lines

incomplete_list = [data_list[i] for i in incomplete_index]
incomplete_list
str_obj = str_split(incomplete_list[4])
print(str_obj)
```

```python
# Return object portion that needs to be completed. 
def incomplete_check(str_obj):
    # Loop through each character object
    check_list = []
    for i in range(len(str_obj)):
        
        # Track characters tested
        check_list.append(str_obj[i])

        # Check if item is open or close
        # If open go to next iteration
        if valid_open.count(str_obj[i]) > 0:
            continue

        else:
            
            # If a close character check if it closes a chunk
            if valid_dict[str_obj[i]] == check_list[-2]:

                # If yes, remove the last two characters
                del check_list[-1]
                del check_list[-1]

    return(check_list)
```

```python
print(incomplete_check(str_obj))
list(reversed(incomplete_check(str_obj)))
```

```python
# Loop through each object in the incomplete list
score_list = []
for q in range(len(incomplete_list)):

    # Gather ending characters needed
    str_obj = str_split(incomplete_list[q])
    need_close = list(reversed(incomplete_check(str_obj)))
    score = 0


    # Loop through each close object
    for m in range(len(need_close)):

        # Calculate the penalty for each character & update score
        penalty_val = close_dict[need_close[m]]
        score = (score*5) + penalty_val
    
    score_list.append(score)
```
