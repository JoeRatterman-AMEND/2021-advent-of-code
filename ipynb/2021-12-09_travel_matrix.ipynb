```python
# Define function to check if index is valid 
def is_valid(array, index):    
    row, column = index
    return 0 <= row < len(array) and 0 <= column < len(array[row])
```

```python
def get_next(node, offset):
    """Find the next location based on an offset from the current location."""
    row, column = node
    row_offset, column_offset = offset
    return row + row_offset, column + column_offset
```

```python
def check_neighbors(array, neighbor_list, test_set):
    next_locs = []
    for n in neighbor_list:

        next_spot = get_next(test_set, n)

        if is_valid(array, next_spot) and int(array[next_spot]) < 9:

            next_locs.append(next_spot)
            
    return(next_locs)
```

```python
basin_size = []
for lp in low_point_index:


    full_basin = []
    tested_spot = []
    level_check = check_neighbors(input_matrix, pair_check, lp)

    for i in level_check:

        next_check = []
        next_check.append(i)

        while len(next_check) > 0:

            #print("check", next_check[0])
            full_basin.append(next_check[0])
            near_neighbs = check_neighbors(input_matrix, pair_check, next_check[0])
            
            next_check.remove(next_check[0])

            
            next_check = next_check + near_neighbs
            next_check = [x for x in next_check if not x in full_basin]


        
    full_basin = list(set(full_basin))
    #print(full_basin)


    size = len(list(set(full_basin)))
    basin_size.append(size)


print(basin_size)
```
