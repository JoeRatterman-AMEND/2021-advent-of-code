```python
# Loaded required packages
from collections import Counter

# Load input data
file_path = "C:\\Users\\JoeRatterman\\Documents\\GitHub\\2021-advent-of-code\\input-files\\2021-12-14_input.txt"
data_list = open(file_path, 'r').read().split('\n')

# Split out original list
beg_str = data_list[0]

# Split out rule updates
update_list = []
for i in range(len(data_list[2:])):
    update_list.append(data_list[2:][i].split(" -> "))
```

```python
# Function to create list of each pair of sequential letters
def generate_str_list(str_obj):

    # Loop through dict and add to dictionary
    init_list = []
    for i in range(len(str_obj)):

        if i + 1 < len(str_obj):

            init_list.append(str_obj[i] + str_obj[i + 1])
    
    return(init_list)

# Function to split string by character
def str_split(word):
    return [char for char in word]
```

```python
# Function to create list of each pair of sequential letters
def generate_str_dict(str_obj):

    # Loop through dict and add to dictionary
    init_dict = {}
    for i in range(len(str_obj)):

        if i + 1 < len(str_obj):

            init_dict[i] = str_obj[i] + str_obj[i + 1]
    
    return(init_dict)
```

```python
# Function to create list of each pair of sequential letters
def generate_str_dict_modified(str_obj):

    # Loop through dict and add to dictionary
    init_dict = {}
    for i in range(len(str_obj)):

        if i + 1 < len(str_obj):

            init_dict[str_obj[i] + str_obj[i + 1]] = init_dict[str_obj[i] + str_obj[i + 1]]
    
    return(init_dict)
```

```python
# Function to map dict values to list
def replace_list(list, dictionary):
    return [dictionary.get(item, item) for item in list]
```
